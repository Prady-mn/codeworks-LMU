```python
def createBasis_Fermions(N = 4,N_up = 2):
    
    states = []
    # --- Gosper's hack ---
    x = (1<<(N_up)) - 1

    limit = 1<<N
    while x < 2**N:
        states.append(x)
        # Gosper's hack to get next integer with same number of 1s
        c = x & -x
        r = x + c
        x = (((r ^ x) >> 2) // c) | r
        
    states.sort()

    #assert len(states) == sp.special.comb(N,N_up)
    return states
```

```python
def hopping(state,index):
    if (state & (1<<index)):
        return state ^ (2<<(index))
    return state
```
