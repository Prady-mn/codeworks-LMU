### Logistic Map

```python
import numpy as np
import matplotlib.pyplot as plt
import typing
from collections import namedtuple
```

```python
import numpy as np

class logisticMap:
    points = None
    iterations = None
    theoretical = None
    cycle = None
    
    def __init__(self, x0=0.1, r=2.8, max_iterations=1000, threshold=1e-6, cycle_count = 4, store_points = False):
        
        self.x0 = x0
        self.r = r
        self.x_initial = x0
        self.max_iterations = max_iterations
        self.threshold = threshold
        self.store_points = store_points
        self.cycle_count = cycle_count
        
        self.theoretical = 1 - 1/r
        self.fixedPoint(x0, r, max_iterations, threshold, cycle_count, store_points)
        
        if(store_points):
            self.points = np.array(self.points)
        else:
            self.points = None

    @staticmethod
    def logistic(x, r):
        return r * x * (1 - x)

    def fixedPoint(self, x0=0.1, r=2.8, max_iterations=1000, threshold=1e-6, cycle_count = 4, store_points=False):

        self.cycle = 1
        self.points = [x0]
        decimals = int(-np.log10(threshold))
        x_next = x0
        
        for i in range(max_iterations):
            
            x_next = r * self.points[-1] * (1 - self.points[-1])
            x_next = round(x_next,decimals)
            
            

            if x_next in self.points[:]:
                if (self.points.count(x_next) >= cycle_count):
                    #print(self.cycle,cycle_count,self.points,len(self.points) -self.points.index(x_next))
                    
                    self.cycle = int((len(self.points) - self.points.index(x_next)) / cycle_count)
                    self.iterations = i + 1
                    self.x0 = self.points[-self.cycle:]
                    
                    return
                
            self.points.append(x_next)

        #IF UNABLE TO FIND CYCLE
        self.iterations = max_iterations
        self.cycle = -1
        self.x0 = self.points[int(0.99*len(self.points)):]
        
        return
    
    def __repr__(self):
        return (
            "LogisticMap(\n"
            f"  x0 = {self.x0},\n"
            f"  cycle = {self.cycle}"
            f"  r = {self.r},\n"
            f"  Iterations = {self.iterations},\n"
            f"  Threshold = {self.threshold},\n"
            f"  Cycle = {self.cycle},\n"
            f"  points = {self.points.shape if self.points is not None else None}\n"
            f"  Theoretical = {self.theoretical}"
            ")"
        )
    def theoretical(r):
        return 1 - 1/r
```

```python
logisticMap(x0 = 0.1,r = 2.8)
```

```python
logisticMap(x0 = 0.4,r = 3.5699,max_iterations=2000,store_points=True)
```

# (a)

```python
%matplotlib widget

def printConvergeceMaps(data, max_iterations=1000, threshold=1e-6,cycle_count = 4):
    maps = [logisticMap(i,j,max_iterations, threshold,cycle_count, store_points=True) for i,j in data]
    
    fig = plt.figure()

    for map in maps:
        line, = plt.plot(list(range(map.iterations)),map.points, label = f'{np.around(map.x0,3)}')
        plt.axhline(map.theoretical,linestyle = "--",color = line.get_color(),alpha = 0.7)

    plt.grid()
    plt.legend(title = "Fixed Point Values")
    plt.xlabel("$n$")
    plt.ylabel("$x_{n}$")

    return fig

plt.close('all')
printConvergeceMaps(data = [(0.5,2.8),(0.5,2.7),(0.5,2.9)])
plt.show()
```
